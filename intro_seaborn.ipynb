# **Introducción a Seaborn**

## ¿Qué es Seaborn?

Seaborn es una biblioteca para crear gráficos estadísticos en Python. Está basada en Matplotlib, y se integra con las estructuras de Pandas. Esta biblioteca es tan potente como Matplotlib, pero aporta simplicidad y funciones inéditas. Permite explorar y comprender rápidamente los datos.

Link: [https://seaborn.pydata.org/index.html](https://seaborn.pydata.org/index.html)

## Como usar e instalar Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

## Datos

```python
import numpy as np
import pandas as pd

heights = np.array([1.3, 1.68, 1.1, 1.92, 1.69])
weights = np.array([65.4, 59.3, 78.1, 95.3, 68.7])
gender = np.array(["H", "M", "M", "H", "NE"])
data = zip(heights, weights, gender)

df = pd.DataFrame(data, columns=['height', 'weight', 'gender'])
```

## Creando primer grafica con seaborn

```python
# Con Matplotlib
fig, ax = plt.subplots()

ax.scatter(heights, weights)

plt.show()
```

```python
# Con Seaborn
sns.scatterplot(data=df, x='height', y='weight')

plt.show()
```

```python
# Con Seaborn
sns.scatterplot(data=df, x='height', y='weight', hue='gender')

plt.show()
```

## RelPlot (Grafico relacional)

Permite visualizar la relación entre 2 variables cuantitativas usando el diagrama de dispersión o de lineas

```python
# Cargar nuevos datos
df_salarios = pd.read_csv('DATASETS/Salario_data.csv')
print(df_salarios.shape)
print(df_salarios.dtypes)
```

```python
# Creando nuestro primer relplot
sns.relplot(data=df_salarios, x='age', y='income', kind='scatter')
plt.show()
```

```python
# Usando col
sns.relplot(data=df_salarios,
            x='age',
            y='income',
            kind='scatter',
            col="education")
plt.show()
```

```python
# Usando row
sns.relplot(data=df_salarios,
            x='age',
            y='income',
            kind='scatter',
            row="education")
plt.show()
```

```python
# Usando row & col
sns.relplot(data=df_salarios,
            x='age',
            y='income',
            kind='scatter',
            col="education",
            row='jobtitle')
plt.show()
```

## Customizando

```python
# Usando size
sns.relplot(data=df_salarios,
            x='age',
            y='income',
            kind='scatter',
            size="jobtitle")
plt.show()
```

```python
# Usando hue
sns.relplot(data=df_salarios,
            x='age',
            y='income',
            kind='scatter',
            size="jobtitle",
            hue='jobtitle')
plt.show()
```

```python
# Usando style
sns.relplot(data=df_salarios,
            x='age',
            y='income',
            kind='scatter',
            size="jobtitle",
            hue="jobtitle",
            style='jobtitle')
plt.show()
```

```python
# Usando alpha
sns.relplot(data=df_salarios,
            x='age',
            y='income',
            kind='scatter',
            size="jobtitle",
            hue="jobtitle",
            style='jobtitle',
            alpha=0.95)
plt.show()
```

## Cambiando el estilo y color

```python
sns.set_style("whitegrid")

sns.relplot(data=df_salarios,
            x='age',
            y='income',
            kind='scatter',
            size="jobtitle",
            hue="jobtitle",
            style='jobtitle',
            alpha=0.4)
plt.show()
```

```python
# Cambiando la paleta de colores
sns.set_palette("RdBu")

sns.relplot(data=df_salarios,
            x='age',
            y='income',
            kind='scatter',
            size="jobtitle",
            style='jobtitle',
            alpha=0.4)
plt.show()
```

```python
# Creando nuestra propia paleta de colores
mis_colores = ["blue", "green", "orange", "purple", "darkred", "darkblue"]

# Usando los colores custom
sns.set_palette(mis_colores)

sns.relplot(data=df_salarios,
            x='age',
            y='income',
            kind='scatter',
            size="jobtitle",
            style='jobtitle',
            alpha=0.4)
plt.show()
```

```python
# Cambiando la escala
# paper, notebook, talk, poster
sns.set_context("talk") # paper es default

# Usando alpha
sns.relplot(data=df_salarios,
            x='age',
            y='income',
            kind='scatter',
            size="jobtitle",
            hue="jobtitle",
            style='jobtitle',
            alpha=0.4)
plt.show()
```

## Titulos y etiquetas

```python
# FacetGrid vs AxesSubplot
g = sns.scatterplot(data=df_salarios,
            x='age',
            y='income')
type(g)
```

```python
g = sns.relplot(data=df_salarios,
            x='age',
            y='income',
            kind='scatter',
            size="jobtitle",
            hue="jobtitle",
            style='jobtitle',
            alpha=0.4)

type(g)
```

```python
g = sns.relplot(data=df_salarios,
            x='age',
            y='income',
            kind='scatter',
            size="jobtitle",
            hue="jobtitle",
            style='jobtitle',
            alpha=0.4)

g.fig.suptitle("Mi titulo", y=1.03) #, y=1.03) #Valor default es 1

plt.show()
```

```python
# Agregando titulo a un AxesSubplot

g = sns.scatterplot(data=df_salarios,
            x='age',
            y='income')

g.set_title("Mi Titulo 2", y=1.03)
g.set(xlabel="edad", ylabel="salario")

plt.show()
```

```python
# Modificando los titulos de los subelementos de un FaceGrid

g = sns.relplot(data=df_salarios,
            x='age',
            y='income',
            kind='scatter',
            row="education")

g.set_titles("Esta es la columna {row_name}")

plt.show()
```
